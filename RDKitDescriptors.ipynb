```python
import deepchem as dc
from rdkit import Chem
import pandas as pd
import numpy as np
import pickle
```

```python
df_smiles = pd.read_csv('data/pretest/smiles_pretest.csv')
```

```python
df_smiles
```

```python
smiles = df_smiles['Smiles']
featurizer = featurizer = dc.feat.RDKitDescriptors()
features = featurizer.featurize(smiles)
```

```python
nan_idx = []
for idx, i in enumerate(features):
    if len(i)==0:
        nan_idx.append(idx)
print(len(nan_idx))
```

```python
features[nan_idx] = np.NaN
```

```python
pickle.dump(features, open('data/pretest/rdkit.pkl', 'wb'))
```
