```python
import pyIsoP.histo as histo
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import pyIsoP.grid3D as grid3D
import pyIsoP.potentials as potentials
import pyIsoP.forcefields as forcefields
import pyIsoP.writer as writer
from ase.atoms import string2symbols
```

```python
def preprocess(path):
    t1=grid3D.grid3D(path, spacing=0.5)          # Intialize grid3D object
    f1=forcefields.forcefields(t1, forcefield='data/UFF', sigma=3.95, epsilon=46)      # Update the force field details to grid obj. t1
    t1= grid3D.grid3D.grid_calc(t1,"lj",f1)
    h = histo.histo(12)                            # Overwrite the existing object with computed 3D grid.
    h = histo.histo.grid2histo(t1, h) 
    return h.RhoE
```

```python
df = pd.read_csv('data/train/train.csv')

filenames = df.MOFname.values
```

```python
xyz_path = 'data/train/mof_cif_train'
```

```python
lis = []

for filename in filenames:
    path = os.path.join(xyz_path, '{}.cif'.format(filename))
    lis.append(preprocess(path))
```
