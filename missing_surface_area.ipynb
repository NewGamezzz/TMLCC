```python
import pandas as pd
import numpy as np
import pickle
from utils.utils import *
from utils.surface_area import *
```

```python
df_train = pd.read_csv('data/train/train.csv')
df_train = df_train.dropna(subset=['functional_groups'])
df_train = df_train.reset_index(drop=True)
df_train.insert(loc=3, column="density [g/cm^3]", value=(df_train["weight [u]"] / df_train["volume [A^3]"]) * 1.66054)
```

```python
miss_idx = np.arange(len(df_train), dtype=int)[df_train['surface_area [m^2/g]']<=0]
surface_area = pickle.load(open("data/train/missing_surface_area.pkl", "rb"))

for i, idx in enumerate(miss_idx):
    if idx in surface_area:
        print(i)
        continue
    density = df_train['density [g/cm^3]'][idx]
    cifname = "data/train/mof_cif_train/{}.cif".format(df_train['MOFname'][idx])

    cell_matrix, inverse_matrix, unit_cell_volume = get_box_matrix(cifname)
    atom_type, pos = get_atomic_coordinates(cifname)

    # GSA
    gsa_sum = 0
    for atom in set(atom_type):
        gsa_ = gsa(atom_type, pos, cell_matrix, inverse_matrix, unit_cell_volume, probe_diameter=3.40, nSample=100, InsertTypeOfAtoms=atom)
        gsa_sum += gsa_


    surface_area[idx] = gsa_sum / density
    pickle.dump(surface_area, open("data/train/missing_surface_area.pkl", "wb"))
    print(i, gsa_sum / density)
```
